```python
import torch
s = torch.tensor(0, dtype=torch.float32)
for i in range(1000):
    s += torch.tensor(0.01, dtype=torch.float32)
print(s)

s = torch.tensor(0, dtype=torch.float16)
for i in range(1000):
    s += torch.tensor(0.01, dtype=torch.float16) # 小sum小因子
print(s)

s = torch.tensor(0, dtype=torch.float32)
for i in range(1000):
    s += torch.tensor(0.01, dtype=torch.float16)  #大sum小因子
print(s)

s = torch.tensor(0, dtype=torch.float32)
for i in range(1000):
    x = torch.tensor(0.01, dtype=torch.float16)
    s += x.type(torch.float32) # 小因子to 大因子 大sum
print(s)
```

```python
import torch
import torch.nn as nn
class ToyModel(nn.Module):
    def __init__(self, in_features: int, out_features: int):
        super().__init__()
        self.fc1 = nn.Linear(in_features, 10, bias=False)
        self.ln = nn.LayerNorm(10)
        self.fc2 = nn.Linear(10, out_features, bias=False)
        self.relu = nn.ReLU()

    def forward(self, x: torch.Tensor):
        x1 = self.fc1(x)
        print(x1.dtype)
        x2 = self.relu(x1)
        print(x2.dtype)
        x3 = self.ln(x2)
        print(x3.dtype)
        x4 = self.fc2(x3)
        print(x4.dtype)
        return x4
    
IN_DIM = 5      # 对应 model 的 in_features
OUT_DIM = 2     # 对应 model 的 out_features
BATCH_SIZE = 3  # 批次大小 (可以是任意整数)

# 3. 构造满足条件的输入 x
# 关键点：第二个维度必须是 IN_DIM (这里是 5)
device: str = "mps" if torch.backends.mps.is_available() else None
x: torch.Tensor = torch.randn(BATCH_SIZE, IN_DIM).to(device = device)
model: torch.nn.Module = ToyModel(in_features=IN_DIM, out_features=OUT_DIM).to(device)
dtype: torch.dtype = torch.float32

with torch.autocast(dtype=torch.bfloat16, device_type=device):
    y = model(x)
    
```
